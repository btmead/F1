```python
import matplotlib.pyplot as plt
import fastf1.plotting
```

```python
fastf1.Cache.enable_cache('C:/Users/benme/OneDrive - The Perse School/Formula 1')
fastf1.plotting.setup_mpl()

session = fastf1.get_session(2021, 'Bahrain', 'Q')
session.load()
```

```python
ver_lap = session.laps.pick_driver('VER').pick_fastest()
ham_lap = session.laps.pick_driver('HAM').pick_fastest()

ver_tel = ver_lap.get_car_data().add_distance()
ham_tel = ham_lap.get_car_data().add_distance()

rbr_color = fastf1.plotting.team_color('RBR')
mer_color = fastf1.plotting.team_color('MER')
```

```python
plt.rcParams['figure.figsize'] = [20, 15]
fig, ax = plt.subplots(3, gridspec_kw={'height_ratios': [2, 1, 1]})

ax[0].plot(ver_tel['Distance'], ver_tel['Speed'], color=rbr_color, label='VER')
ax[0].plot(ham_tel['Distance'], ham_tel['Speed'], color=mer_color, label='HAM')
ax[0].set_ylabel('Speed in km/h')


ax[1].plot(ver_tel['Distance'], ver_tel['Throttle'], color=rbr_color, label='VER')
ax[1].plot(ham_tel['Distance'], ham_tel['Throttle'], color=mer_color, label='HAM')
ax[1].set_ylabel('Throttle Percentage')

ax[2].plot(ver_tel['Distance'], ver_tel['nGear'], color=rbr_color, label='VER')
ax[2].plot(ham_tel['Distance'], ham_tel['nGear'], color=mer_color, label='HAM')
ax[2].set_ylabel('Gear')


plt.show()
```
