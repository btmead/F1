```python
import matplotlib.pyplot as plt
import fastf1.plotting

fastf1.Cache.enable_cache('C:/Users/benme/OneDrive - The Perse School/Formula 1')
fastf1.plotting.setup_mpl()

session = fastf1.get_session(2021, 'Bahrain', 'R')
session.load()

ver_laps = session.laps.pick_driver('VER')
ham_laps = session.laps.pick_driver('HAM')




rbr_color = fastf1.plotting.team_color('RBR')
mer_color = fastf1.plotting.team_color('MER')
```

```python
fig, ax = plt.subplots()

ax.plot(ver_laps['LapNumber'], ver_laps['LapTime'], color=rbr_color, label='VER')
ax.plot(ham_laps['LapNumber'], ham_laps['LapTime'], color=mer_color, label='HAM')
ax.set_ylabel('Lap Time')
ax = plt.gca()
ax.set_ylim([0.00107638888888889, 0.00122685185185185])
ax.set_xlabel('Lap Number')



plt.show()
```
